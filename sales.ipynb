Further Analysis with the help of Visualizations

```python
import seaborn as sns
```

```python
#distribution of ratings
sns.countplot(x='rating', data=df)
```

```python
#distibution od sales by year
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
sns.countplot(x='year', data=df)
```

```python
#you can see that 2015 is the best year of sales
```

```python
#what brand is most saled in 2015 then we check in 2016 

plt.figure(figsize=(12, 6))
sns.countplot(x='brand', data=df[df['year'] == 2015], order=df['brand'].value_counts().iloc[1:10].index)
plt.title("2015")

plt.figure(figsize=(12, 6))
sns.countplot(x='brand', data=df[df['year'] == 2016], order=df['brand'].value_counts().iloc[1:10].index)
plt.title("2016")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(12, 6))

# Get the counts of each brand for the year 2016
brand_counts = df[df['year'] == 2015]['brand'].value_counts()

# Get the top 2nd to 10th most frequent brands and sort them in ascending order
top_brands = brand_counts.iloc[1:10].sort_values(ascending=False).index

# Create the count plot with bars in ascending order
sns.countplot(
    x='brand', 
    data=df[df['year'] == 2015], 
    order=top_brands
)

# Add the title
plt.title(' Sales in 2015 ')

# Show the plot
plt.show()
```

```python
#we can see that bose, logitech and mpow is top sales in 2015
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(12, 6))

# Get the counts of each brand for the year 2016
brand_counts = df[df['year'] == 2016]['brand'].value_counts()

# Get the top 2nd to 10th most frequent brands and sort them in ascending order
top_brands = brand_counts.iloc[1:10].sort_values(ascending=False).index

# Create the count plot with bars in ascending order
sns.countplot(
    x='brand', 
    data=df[df['year'] == 2016], 
    order=top_brands
)

# Add the title
plt.title(' Sales in 2016 ')

# Show the plot
plt.show()
```

```python
#we can see that bose, logitech and TaoTronics is top sales in 2015
```

```python
#From above analysis we can know that january has most sales  
```

```python
#lets see what products were sold the most in january 
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(20, 10))

# Get the counts of each category for the month of January
category_counts = df[df['month'] == 1]['category'].value_counts()

# Get the top 2nd to 10th most frequent categories (already in descending order)
top_categories = category_counts.iloc[1:10].index

sns.countplot(
    x='category', 
    data=df[df['month'] == 1], 
    order=top_categories
)

plt.title('Distribution of Top products in January')

plt.show()
```

```python
#we can see that computers & Accessories are have most sales then camera & photo
```
